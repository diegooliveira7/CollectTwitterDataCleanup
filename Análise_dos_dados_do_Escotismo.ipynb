```python
import os
import dotenv
import requests
```

```python
dotenv.load_dotenv()

def analise_sentimento(text: str) -> str:
  modelo_engine = "text-davinci-003"
  comando = f"Responda em uma única palavra, sendo positivo ou negativo o sentimento contido no seguinte texto: '{text}'"

  cabecalho = {
      "Content-Type": "application/json",
      "Authorization": f"Bearer {'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'}",
  }

  dados = {
      "prompt": comando,
      "temperature": 0.7,
      "max_tokens": 35,
      "n": 1,
      "stop": None,
  }

  reposta = requests.post(
      f"https://api.openai.com/v1/engines/{modelo_engine}/completions",
      headers=cabecalho,
      json=dados,
 )

  return reposta.json()["choices"][0]["text"].strip().lower()
```

```python
analise_sentimento("Chato")
```

```python
dfClass['GPT'] = 1
count = 0

for tweet in dfClass['Tweet']:
  tweet_status = analise_sentimento((tweet))

  if (tweet_status == 'positivo.'):
    dfClass['GPT'][count] = 1
  elif (tweet_status == 'negativo.'):
    dfClass['GPT'][count] = 0
  else:
    dfClass['GPT'][count] = 2
  count = count + 1
```

```python
count = 0
countt = 0
while (countt < 100):
  if (dfClass.iloc[countt]['Binário'] == dfClass.iloc[countt]['GPT']):
    count = count + 1
  countt = countt + 1

count
```
