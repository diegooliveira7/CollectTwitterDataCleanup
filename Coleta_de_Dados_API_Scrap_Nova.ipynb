```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
#Escoteiro until:2022-09-30 since:2022-01-01
query = "(Escoteiro OR Escotismo OR Desbravadores OR Escoteiros OR escotismo OR desbravador OR dbv OR Campori) lang:pt until:2023-05-05 since:2020-01-01"
tweets = []
limit = 100000

for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    
    # print(vars(tweet))
    # break
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.username, tweet.content, tweet.id])
        print(tweet.content)
        
df = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet', 'ID'])
```
