#Classificação

```python
!pip3 install transformers
from transformers import pipeline
##https://https://huggingface.co/Chirayu/mt5-multilingual-sentiment
unmasker003 = pipeline(model='Chirayu/mt5-multilingual-sentiment')
unmasker003("Feliz aniversário")
```

```python
dfClass['Class'] = 1
count = 0

for tweet in dfClass['Tweet']:
  tweet_status = unmasker003(str(tweet))

  if (tweet_status[0]['generated_text'] == 'Positive'):
    dfClass['Class'][count] = 1
  elif (tweet_status[0]['generated_text'] == 'Negative'):
    dfClass['Class'][count] = 0
  else:
    dfClass['Class'][count] = 2
  count = count + 1
```
