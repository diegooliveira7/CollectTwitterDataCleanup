```python
##https://huggingface.co/philschmid/distilbert-base-multilingual-cased-sentiment-2?text=I+like+you.+I+love+you
from transformers import pipeline
unmasker = pipeline(model='philschmid/distilbert-base-multilingual-cased-sentiment-2')
```

```python
dfClass['teste001'] = 1 #Adicionando uma coluna a mais
count = 0
for tweet in dfClass['Tweet']:
  tweet_status = unmasker(str(tweet))

  if (tweet_status[0]['label'] == 'positive'):
    dfClass['teste001'][count] = 1
  elif (tweet_status[0]['label'] == 'negative'):
    dfClass['teste001'][count] = 0
  else:
    dfClass['teste001'][count] = 2

  #dfClass['teste001'][count] = tweet_status[0]['label'].eguals('')
  count = count + 1
```

```python
count = 0
countt = 0
while (countt < 1000):
  if (dfClass.iloc[countt]['Binário'] == dfClass.iloc[countt]['teste001']):
    count = count + 1
  countt = countt + 1

count
```
